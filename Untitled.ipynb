```python
def read_lexicon(lex_path):
    lexicon = {}
    with open(lex_path) as f:
        for line in f:
            temp = re.split(r"\s+", line.strip("\n"))
            word = temp[0]
            phones = temp[1:]
            if word.lower() not in lexicon:
                lexicon[word.lower()] = phones
    return lexicon
```

```python
def get_words_to_phonemes(textgrid,f):
    
    sil_phones = ["sil", "sp", "spn"]
    w_tier = textgrid.get_tier_by_name("words")
    p_tier = textgrid.get_tier_by_name("phones")
    phones = []
    for t in p_tier._objects:
        if phones == []:
            if t.text in sil_phones:
                continue
        phones.append(t)
    words = []
    idx = 0
#     print(len(phones))
    for t in w_tier._objects:
        count = idx
        if words == []:
            if t.text in sil_phones:
                continue
        s, e = t.start_time,t.end_time
        if (phones[idx].text in sil_phones) and s != phones[idx].start_time:
            idx+=1
        if s == phones[idx].start_time:
            while e != phones[count].end_time:
                count+=1
            idx = count+1
            words.append(idx)
#             print("done",count)
        else:
            print("start not found",f)
            return None
    new_words = []
    new_words.append(words[0])
    for i in range(1,len(words)):
        new_words.append(words[i]-words[i-1])
        
    return new_words
        
```
